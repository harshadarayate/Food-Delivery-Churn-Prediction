```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()

def generate_food_delivery_churn_dataset(n=500):
    data = []
    for i in range(101, 101 + n):
        customer_id = i
        orders_last_3_months = np.random.poisson(lam=5)
        avg_order_value = f"₹{random.randint(150, 600)}"
        app_opens_per_week = np.random.randint(1, 15)
        complaints = np.random.poisson(lam=1)
        used_offers = random.choice(["Yes", "No"])
        last_order_date = fake.date_between(start_date='-1y', end_date='today')
        membership = random.choice(["Yes", "No"])
        is_churned = int((orders_last_3_months <= 2 and app_opens_per_week <= 3) or complaints >= 3)

        data.append([
            customer_id,
            orders_last_3_months,
            avg_order_value,
            app_opens_per_week,
            complaints,
            used_offers,
            last_order_date,
            membership,
            is_churned
        ])
    
    columns = [
        "Customer_ID", "Orders_Last_3_Months", "Avg_Order_Value", "App_Opens_Per_Week",
        "Complaints", "Used_Offers", "Last_Order_Date", "Membership", "Is_Churned"
    ]
    
    return pd.DataFrame(data, columns=columns)

# Create and display the dataset
churn_df = generate_food_delivery_churn_dataset(500)

```

```python
churn_df.to_csv('./food_delivery_churn_dataset.csv', index=False)
```
